```python
from cmdstanpy import CmdStanModel
import pandas as pd
import arviz as az 
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

## Excercise 1 - Divergence problem

### Variant A

```python
model = CmdStanModel(stan_file='stan1.stan')
```

```python
result = model.sample(data={'N':1, 'y':[1]}, 
                            seed = 9012022,
                            chains = 4)
```

```python
print(result.diagnose())
```

```python
#Convertion to arviz InferenceData
arviz_result = az.from_cmdstanpy(
    posterior=result
)
#Calculating log sigma
post = arviz_result.posterior
post["log_sigma"] = np.log(post["sigma"])
arviz_result
```

```python
#Arviz plot
az.plot_pair(arviz_result, var_names=['mu', 'log_sigma'], divergences=True)
```

## Excercise 3 - Predictive checks

```python
#Data read from normal.csv
df = pd.read_csv("https://raw.githubusercontent.com/KAIR-ISZ/public_lectures/master/Data%20Analytics%202022/Lab%203%20-%20Probability%20distribution/normal.csv", index_col=0)
df.head()
y = df["value"] 
N = len(y)
```

### Posterior predictive

```python
model = CmdStanModel(stan_file='stan3.stan')
```

```python
result = model.sample(data={'N':N, 'y':y}, 
                            seed = 9012022,
                            chains = 4)
```

```python
result.summary()
```

```python
#Convertion to arviz InferenceData
arviz_result = az.from_cmdstanpy(
    posterior=result,
    posterior_predictive=["y_rep"], 
    observed_data={"y": y},
)
arviz_result
```

```python
#Arviz plot
az.plot_ppc(arviz_result,data_pairs={"y": "y_rep"})
```

### Prior predictive

```python
model = CmdStanModel(stan_file='stan4.stan')
```

```python
result = model.sample(data={'N':100}, 
                            chains = 1,
                            fixed_param=True,
                            seed = 523833
                            )
```

```python
result.summary()
```

```python
#Convertion to arviz InferenceData
arviz_result = az.from_cmdstanpy(
    prior=result,
    prior_predictive=["y_prior"],
    observed_data={"y": y}
)
arviz_result
```

```python
az.plot_ppc(arviz_result, group="prior", data_pairs={"y": "y_prior"}, num_pp_samples=30)
plt.axis([-5, 4, 0, 1])
```

```python
#Draw y_prior histogram
y_prior = result.draws_pd().drop(columns=["lp__", "accept_stat__", "mu", "sigma"])
y_prior["y_prior[1]"].hist()
plt.title("y_prior[1]")
plt.show()
y_prior["y_prior[25]"].hist()
plt.title("y_prior[25]")
plt.show()
y_prior["y_prior[50]"].hist()
plt.title("y_prior[50]")
plt.show()
y_prior["y_prior[100]"].hist()
plt.title("y_prior[100]")
plt.show()
```
